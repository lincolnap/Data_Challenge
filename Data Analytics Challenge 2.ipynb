```python
#Importacion de las librerias 
import tweepy
import Api_keys #archivo de las credenciales
import csv
```

```python
# Metodo de autenticacion OAuth 2 (read only)
consumer_key = Api_keys.key1
consumer_secret = Api_keys.key2

# Metodo de autenticacion OAuth
auth = tweepy.OAuthHandler( consumer_key, consumer_secret )

api = tweepy.API( auth )

#Variable para hacer la busqueda en de las palabras en twiieter
search = 'Panama'

#Escritura de archivo csv

# cracion del archivo en donde se alamacenara las datos para analizarlos en excel
csvFile = open('result_final-test.csv', 'a', newline='',encoding="utf-8")

#Escritos de csv
csvWriter = csv.writer(csvFile)
```

```python
#Envio de parametros y obtencion de los tweet
for busqueda in tweepy.Cursor( api.search, q=search, lang='es', busqueda_mode='extended').items( 1000 ):
    csvWriter.writerow([busqueda.created_at, busqueda.user.screen_name, busqueda.user.name, busqueda.text.encode('utf-8'), busqueda.user.name])
    
    print( busqueda.created_at, busqueda.user.screen_name, busqueda.user.name,
           busqueda.text, busqueda.user.name)
```
