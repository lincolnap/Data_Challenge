```python
# AUTENTICACION GET, METODO 2 MANUAL CODE and save Json

# Importar librerias 
import io
import requests
import Api_keys #Codigo para guardar las Claves 
import json
```

```python
#LLamado a la API de Tweeter y obtencion de los datos

def auth_token():
    BEARER_TOKEN = Api_keys.key_token
    return BEARER_TOKEN


# construction de la consulta URL a obtener
def create_url():
    busqueda = "panama"
    query = "query="
    max = 100
    max_results = "max_results=" + str( max )
    fields = "tweet.fields=author_id,text,id,created_at,source"
    url = f"https://api.twitter.com/2/tweets/search/recent?{query + busqueda}&{max_results}&{fields}"
    return url


def cabezera_principal(bearer_token):
    headers = {"Authorization": "Bearer {}".format( bearer_token )}
    return headers


def conecxion(url, headers):
    response = requests.request( "GET", url, headers=headers )
    # print( response.status_code )
    if response.status_code != 200:
        raise Exception(
            "Request returned an error: {} {}".format(
                response.status_code, response.text
            )
        )
    return response.json()

#Llamado a la API
def main(url):
    bearer_token = auth_token()
    headers = cabezera_principal( bearer_token )
    response = conecxion( url, headers )

    return response
```

```python
# Almacenamiento del Token de paginacion del API
Next_token1 = json.dumps( main( create_url() )['meta']['next_token'] )


'''Funcionamiento para el ingreso de los token de paginacion proporcionados por Twitter para las demas valores
   Nota: Se creo esta funcion similar para ver los parametros enviados para la paginas. Se puede simpificar en uno solo.'''

def create_url2(ing_next_token):
    busqueda = "Panama"
    query = "query="
    max = 100
    max_results = "max_results=" + str( max )
    # field ="tweet.fields=author_id, context_annotations, conversation_id, created_at, entities, geo, id, in_reply_to_user_id, lang, non_public_metrics, organic_metrics, possibly_sensitive, promoted_metrics, public_metrics, referenced_tweets, source, text"
    fields = "tweet.fields=author_id,text,id,created_at,source"
    next_token = "next_token=" + str( ing_next_token ).strip( '"' )
    url2 = f"https://api.twitter.com/2/tweets/search/recent?{query + busqueda}&{max_results}&{next_token}&{fields}"
    return url2
```

```python
# Variables Para el enviar el token al token
Next_token2 = json.dumps( main( create_url2( Next_token1 ) )['meta']['next_token'] )
Next_token3 = json.dumps( main( create_url2( Next_token2 ) )['meta']['next_token'] )
Next_token4 = json.dumps( main( create_url2( Next_token3 ) )['meta']['next_token'] )
Next_token5 = json.dumps( main( create_url2( Next_token4 ) )['meta']['next_token'] )
Next_token6 = json.dumps( main( create_url2( Next_token4 ) )['meta']['next_token'] )
Next_token7 = json.dumps( main( create_url2( Next_token4 ) )['meta']['next_token'] )
Next_token8 = json.dumps( main( create_url2( Next_token4 ) )['meta']['next_token'] )
Next_token9 = json.dumps( main( create_url2( Next_token4 ) )['meta']['next_token'] )
Next_token10 = json.dumps( main( create_url2( Next_token4 ) )['meta']['next_token'] )

# Informacion de los tweet de las paginas solicitadas

Pags = main( create_url() )['data'], main( create_url2( Next_token1 ) )['data'], \
       main( create_url2( Next_token2 ) )['data'], main( create_url2( Next_token3 ) )['data'], \
       main( create_url2( Next_token4 ) )['data'], main( create_url2( Next_token5 ) )['data'], \
       main( create_url2( Next_token6 ) )['data'], main( create_url2( Next_token7 ) )['data'], \
       main( create_url2( Next_token8 ) )['data'], main( create_url2( Next_token8 ) )['data'], \
       main( create_url2( Next_token9 ) )['data'], main( create_url2( Next_token10 ) )['data']
```

```python
#Variable para union en listas en un diccionario Json
Tweets = {'tweets_1000': Pags}
#Guardar valores en un archivo Json 

with open( '1000_Tweets.json', 'w', encoding='utf8' ) as file:
    json.dump( Tweets, file, indent=4, ensure_ascii=False )

#Imprimir en consola la informacion guardada

print( json.dumps( Pags, indent=4, ensure_ascii=False ) )
```
