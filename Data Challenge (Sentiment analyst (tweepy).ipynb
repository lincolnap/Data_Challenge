## Limpieza de caracteres no deseados

```python
def limpieza_tweets(tweets):
    tweets = re.sub(r'^RT[\s]+', '', tweets)
    tweets = re.sub(r':', '', tweets)
    tweets = re.sub(r'https?:\/\/.*[\r\n]*', '', tweets)
    tweets = re.sub(r'#', '', tweets)
    tweets = re.sub(r'@[A-Za-z0-9]+', '', tweets)
    return tweets
```

```python
df['Tweets texts'] = df['full_text'].apply(limpieza_tweets)
df['Tweets texts']
```

<h3> Analisis de sentimiento con la libreria de TextBlob<\h3>

## Defir las variable de la polaridad

```python
def polaridad(tweet):
    analysis = TextBlob(tweet)
    if tweet != '':
        if analysis.detect_language() == 'es':
            result = analysis.translate(from_lang = 'es', to = 'en').sentiment.polarity
            time.sleep(3)
            return result

def subjetividad(tweet):
    analysis = TextBlob(tweet)
    if tweet != '':
        if analysis.detect_language() == 'es':
            result = analysis.translate(from_lang = 'es', to = 'en').sentiment.subjectivity
            time.sleep(3)
            return result
```

## Aplicando las funciones para saber el score de ellas

```python
df['polaridad'] = df['Tweets texts'].apply(polaridad)
df['subjetividad'] = df['Tweets texts'].apply(subjetividad)

df[['Tweets texts','polaridad','subjetividad']]
```

```python
df.to_csv(r'Analisis de Sentemientos searh Panama.csv',columns=['Tweets texts','polaridad','subjetividad'])
```
